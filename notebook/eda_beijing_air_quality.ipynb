# Import Libraries

```python
import os
import re
import pandas as pd
import matplotlib.pyplot as plt
```

# Read Data

```python
path = "../data/raw"

csv_files = [file for file in os.listdir(path=path) if file.endswith(".csv")]
csv_files
```

```python
csv_data = {}

for file in csv_files:
    name = re.sub(r'[^a-zA-Z0-9_]', '_', file.replace("PRSA_Data_", "").replace(".csv", ""))
    full_path = os.path.join(path, file)
    
    df = pd.read_csv(full_path, index_col=0)
    print(f"\n{name}")
    display(df.head())
    csv_data[name] = df
```

# Tally Columns

```python
keys = list(csv_data.keys())
ref_df = csv_data[keys[0]]
ref_cols = ref_df.columns
ref_cols
```

```python
for key in keys[1:]:
    df = csv_data[key]

    if not df.columns.equals(ref_cols):
        print("Does not equal.")
```

# Concat dataframe

```python
df = pd.DataFrame()
num_rows = 0

for key in csv_data.keys():
    num_rows += len(csv_data[key])
    df = pd.concat([df, csv_data[key]], axis=0, ignore_index=True)
```

```python
len(df) == num_rows
```
